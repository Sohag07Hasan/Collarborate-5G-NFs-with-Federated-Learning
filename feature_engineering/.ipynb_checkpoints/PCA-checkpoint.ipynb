```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

import tensorflow as tf
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.regularizers import l2
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Dropout
sns.set(font_scale=1)
```

```python
#Data Sources
PATH = './data/ids_sdn/'
training_set = PATH + "Train_ULAK.csv"
test_set = PATH + "Test_ULAK.csv"
training_set_sanitized = PATH + "Train_ULAK_Sanitized.csv"
test_set_sanitized = PATH + "Test_ULAK_Sanitized.csv"
```

```python
train_df = pd.read_csv(training_set_sanitized)
```

```python
#Sanitize the data

#Removing outer spaces from column/feature Names
train_df.columns = train_df.columns.str.strip()
train_df.columns = train_df.columns.str.replace(' ', '_')

#Remove unwanted text from Label Column
train_df['Label'] = train_df['Label'].str.replace('�', '_', regex=True)
train_df['Label'] = train_df['Label'].str.replace(' ', '_', regex=True)
train_df['Label'] = train_df['Label'].str.replace('___', '_', regex=True)
```

```python
# Calculate the number of null values column-wise
null_counts = train_df.isnull().sum()

# Calculate the number of infinite values column-wise
def count_infinite_values(series):
    return sum([1 for value in series if isinstance(value, (float, np.float64)) and np.isinf(value)])

# Calculate the number of infinite values column-wise
#infinite_counts = df.apply(lambda x: np.isinf(x).sum())
infinite_counts = train_df.apply(count_infinite_values)
```

```python
print("Null Counts:")
for key, value in null_counts.items():
    if value != 0:
        print(key, value)


print("\nInfinite Counts:")
for key, value in infinite_counts.items():
    if value != 0:
        print(key, value)
```

```python
#Total Record
print("Total: ",len(train_df['Flow_Duration']))

#Recoreds per Lavel
Label_counts = train_df['Label'].value_counts()
print (Label_counts)
```

```python
#Removing infinity and null values
train_df = train_df.replace([np.inf, -np.inf], np.nan)
train_df = train_df.dropna()
```

```python
##  Applying PCA Algorithms
from sklearn.decomposition import PCA
```

```python
pca_component_count = len(scaled_df.columns)
pca_components = ['pc_{}'.format(i+1) for i in range(pca_component_count)]
#print(pca_components)
```

```python
pca = PCA(n_components = pca_component_count)
data_pca = pca.fit_transform(scaled_data)
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
```

```python
pca_component_count = 50
pca_components = ['pc_{}'.format(i+1) for i in range(pca_component_count)]
#print(pca_components)
pca = PCA(n_components = pca_component_count)
data_pca = pca.fit_transform(scaled_data)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
```

```python
feature_loadings = pd.DataFrame(abs(pca.components_), columns=input_features, index=pca_components)
```

```python
feature_loadings.head()
```

```python
# Calculate the absolute sum of loadings for each feature
feature_importance = feature_loadings.sum(axis=0)
# Rank the features based on their importance
ranked_features = feature_importance.sort_values(ascending=False)
```

```python
# Print the ranked features
print(ranked_features.round(4))
```
